Our dataset contains information on residential properties in Mexico, including housing prices, location details, and property characteristics. It is used to analyze housing trends and price variations across different regions in the Mexican real estate market.

#### Loading the Dataset from the CSV file and Cleaning 

```python
import pandas as pd 
df1=pd.read_csv("C:/Users/USER/Downloads/mexico-real-estate-1.csv")
df2=pd.read_csv("C:/Users/USER/Downloads/mexico-real-estate-2.csv")
df3=pd.read_csv("C:/Users/USER/Downloads/mexico-real-estate-3.csv")
```

```python
df1.shape
df1.info()
print(df1.shape)
df1.head(10)
```

```python
#remove"NaN" values
df1.dropna(inplace=True)
df1.info()
```

```python
#remove"NaN" values
df1.dropna(inplace=True)
# Clean and convert the price column
df1["price_usd"] = (
    df1["price_usd"]
    .astype(str)  # ensure it's string
    .str.replace("$", "", regex=False)
    .str.replace(",", "")
    .astype(float)  # finally convert to float
)
df1.info()
```

```python
df2.shape
df2.info()
df2.head()
```

```python
#remove "NaN" values
df2.dropna(inplace=True)
df2.info()
df2.head()
```

```python
#remove "NaN" values
df2.dropna(inplace=True)
#df2.info()
#create "price_usd" col
df2["price_mxn"].head()
df2["price_usd"]=(df2["price_mxn"]/19).round(2)
df2.head()
```

```python
#remove "NaN" values
df2.dropna(inplace=True)
#df2.info()
#create "price_usd" col
df2["price_mxn"].head()
df2["price_usd"]=(df2["price_mxn"]/19).round(2)
#drop "price_mxn"col
df2.drop(columns=["price_mxn"], inplace=True)
df2.head()
```

```python
df3.shape
df3.info()
df3.head()
```

```python
#drop 'NaN' values
df3.dropna(inplace=True)
df3.head()
#split "lat-lon" col
df3[["lat", "lon"]] = df3["lat-lon"].str.split(",", expand=True)
df3.head()
```

```python
df3.head()
```

```python
#create 'state' col from "place_with_parent_names"
df3['state']=df3["place_with_parent_names"].str.split("|", expand=True)[2]
#drop cols
df3.drop(columns=["place_with_parent_names","lat-lon"], inplace=True)
df3.head()
```

```python
#concatenate the csv files,taking axis is 0
df=pd.concat([df1,df2,df3])
print(df.shape)
df.head()
```

#### Loading the Cleaned Dataset

After completing the data cleaning process, we loaded the cleaned dataset into the Python environment for further analysis and visualization

```python
#saving the cleaned dataframe
df.to_csv("C:/Users/USER/Downloads/mexico data.clean.csv", index=False)
```

##### Boxplot of Home Sizes

The boxplot summarizes the distribution of home sizes in square meters by displaying the minimum, maximum, and key quartiles. The box represents the interquartile range (25th to 75th percentile), showing where the middle 50% of home sizes lie, while the line inside the box indicates the median home size. The whiskers extend to show the overall spread of the data. #### Distribution of Home Prices using a Histogram

The histogram shows the distribution of home prices in U.S. dollars. The x-axis represents the price ranges, while the y-axis indicates the frequency of properties within each range.
```python
#price per square meter
df["price_per_m2"]=df["price_usd"]/df["area_m2"]
df.head()
```

#### Average House Price per Square Meter by State
This bar chart presents the average house price per square meter across states, computed by grouping properties by state and taking the mean. The states are ordered from highest to lowest average price, making regional differences in housing costs easy to compare. The results show substantial variation in price per square meter, indicating that location plays a significant role in determining housing affordability and market value.

```python
#use groupby
(
    df
    .groupby("state") 
    ["price_per_m2"].mean()
    .sort_values(ascending=False)
    .plot(
        kind="bar",
        xlabel="State",
        ylabel="Price[usd]",
        title="Mean House Price by State"
       )
);
```


#### Relationship Between Home Size and Price
This scatter plot visualizes how the price of homes (in USD) varies with their size (in square meters). Each point represents a property. The plot suggests a general trend where larger homes tend to have higher prices, although there is variability, indicating that other factors besides size may influence home prices. Outliers are also visible, which could represent unusually high-priced or low-priced properties for a given area.

```python
#Price vs Area scatter plot to define the relationship between home size and price using matplotly 
plt.scatter(x=df["area_m2"], y=df["price_usd"])
plt.xlabel("Area[sq meters]")
plt.ylabel("Price [USD]")
plt.title("Price vs Area");
```

```python
#representing the correlation numerically (using pearson correlation-coefficient)
p_correlation=df["area_m2"].corr(df["price_usd"])
print(p_correlation)
#hence there is a moderate relationship between house size and price in mexico states
```
